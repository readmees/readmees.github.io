### Vraag 10: Foutafhandeling
Hoe vang je een KeyError op in Python?
1. ```try: ... except: ...```: op deze manier zorg je namelijk dat het voor elke error werkt, dat is netter voor als er ineens een andere error komt
2. ```try: ... except KeyError: ...```: op deze manier zorg je namelijk dat het voor alleen de KeyError werkt, dat is netter voor als er ineens een andere error komt
3. ```if error in ... : ...```: op deze manier voorkom je elke error, dat is netter voor als er ineens een andere error komt
4. ```if KeyError in ... : ...```: op deze manier voorkom je elke error, dat is netter voor als er ineens een andere error komt
