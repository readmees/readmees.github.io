### Opdracht 1: Speciale Methoden in Python

Maak een Class genaamd `Bibliotheek`. De Class moet de volgende attributen hebben:
- `boeken`: Een lijst met boektitels in de bibliotheek.

En de volgende methoden:
- `__len__()`: Geeft het aantal boeken in de bibliotheek terug.
- `__str__()`: Geeft een stringrepresentatie van de bibliotheek terug in de vorm van "Bibliotheek bevat [aantal boeken] boeken."
- `voeg_boek_toe(boek_titel)`: Voegt een boek toe aan de bibliotheek.
- `__getitem__(index)`: Maakt het mogelijk om een boek uit de bibliotheek op te halen via indexering.
- `__sub__(other)`: Deze methode moet een nieuw `Bibliotheek` object teruggeven dat alleen de boeken bevat die aanwezig zijn in de huidige bibliotheek en niet in de `other` bibliotheek.
- `boekaanwezig(naam)`: Deze methode moet het boek teruggeven aan de hand van de naam
**Tips:**
- Gebruik het `in` sleutelwoord om te controleren of een boek in een bibliotheek zit.
- Onthoud dat de speciale methoden met dubbele underscores je in staat stellen om het standaardgedrag van Python's ingebouwde functies of operatoren te overschrijven.

**Voorbeeldgebruik:**

```python
bib1 = Bibliotheek()
bib1.voeg_boek_toe("Python Basics")
bib1.voeg_boek_toe("Geavanceerd Python")

bib2 = Bibliotheek()
bib2.voeg_boek_toe("Geavanceerd Python")
bib2.voeg_boek_toe("Machine Learning")

print(len(bib1))  # Moet 2 printen
print(str(bib1))  # Moet "Bibliotheek bevat 2 boeken." printen
print(bib1[0])    # Moet "Python Basics" printen

unieke_boeken = bib1 - bib2
print(str(unieke_boeken))  # Moet "Bibliotheek bevat 1 boeken." printen
print(unieke_boeken[0])    # Moet "Python Basics" printen
```

````python
class Bibliotheek:
    def __init__(self):
        self.boeken = []

    def voeg_boek_toe(self, boek_titel):
        self.boeken.append(boek_titel)

    def __len__(self):
        return len(self.boeken)

    def __str__(self):
        return f"Bibliotheek bevat {len(self.boeken)} boeken."

    def __getitem__(self, index):
        print('get item')
        return self.boeken[index]

    def __sub__(self, other):
        unieke_boeken = [boek for boek in self.boeken if boek not in other.boeken]
        resultaat = Bibliotheek()
        resultaat.boeken = unieke_boeken
        return resultaat

# Voorbeeldgebruik:
bib1 = Bibliotheek()
bib1.voeg_boek_toe("Python Basics")
bib1.voeg_boek_toe("Geavanceerd Python")

bib2 = Bibliotheek()
bib2.voeg_boek_toe("Geavanceerd Python")
bib2.voeg_boek_toe("Machine Learning")

print(len(bib1))             # Moet 2 printen
print(str(bib1))             # Moet "Bibliotheek bevat 2 boeken." printen
print(bib1[0])               # Moet "Python Basics" printen

unieke_boeken = bib1.__sub__(bib2)

print(unieke_boeken)   # Moet "Bibliotheek bevat 1 boeken." printen
print(unieke_boeken[0])      # Moet "Python Basics" printen
````

### Opdracht 2: Bankrekening Class

Maak een Class genaamd `Bankrekening`. Deze Class moet de volgende eigenschappen hebben:

- `rekeninghouder`: De naam van de persoon die de rekening bezit.
- `saldo`: Het huidige saldo van de rekening. Het begint standaard met een saldo van 0.

En de volgende methoden:

- `storten(bedrag)`: Hiermee kan een bedrag worden gestort op de bankrekening. Het saldo moet dienovereenkomstig worden verhoogd.
- `opnemen(bedrag)`: Hiermee kan een bedrag worden opgenomen van de bankrekening. Als het saldo niet toereikend is, moet een bericht worden afgedrukt dat er onvoldoende saldo is. Anders moet het saldo worden verminderd met het opgenomen bedrag.
- `__str__()`: Geeft een stringrepresentatie van de bankrekening terug in de vorm van "[Rekeninghouder]'s rekening heeft een saldo van [saldo]."

**Voorbeeldgebruik:**

```python
rekening = Bankrekening("Jan de Vries")

print(rekening)   # Moet printen: "Jan de Vries's rekening heeft een saldo van 0."

rekening.storten(500)
print(rekening)   # Moet printen: "Jan de Vries's rekening heeft een saldo van 500."

rekening.opnemen(200)
print(rekening)   # Moet printen: "Jan de Vries's rekening heeft een saldo van 300."

rekening.opnemen(500)   # Moet printen: "Onvoldoende saldo."
```

````python
class Bankrekening:
    def __init__(self, rekeninghouder):
        self.rekeninghouder = rekeninghouder
        self.saldo = 0

    def storten(self, bedrag):
        if bedrag < 0:
            print("Je kunt geen negatief bedrag storten.")
            return
        self.saldo += bedrag

    def opnemen(self, bedrag):
        if bedrag < 0:
            print("Je kunt geen negatief bedrag opnemen.")
            return
        if bedrag > self.saldo:
            print("Onvoldoende saldo.")
        else:
            self.saldo -= bedrag

    def __str__(self):
        return f"{self.rekeninghouder}'s rekening heeft een saldo van {self.saldo}."

# Voorbeeldgebruik:

rekening = Bankrekening("Jan de Vries")

print(rekening)   # "Jan de Vries's rekening heeft een saldo van 0."

rekening.storten(500)
print(rekening)   # "Jan de Vries's rekening heeft een saldo van 500."

rekening.opnemen(200)
print(rekening)   # "Jan de Vries's rekening heeft een saldo van 300."

rekening.opnemen(500)   # "Onvoldoende saldo."
````
