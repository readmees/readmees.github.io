```python
''' Opdracht 4: Afbeeldingen - Nu het is ook mogelijk om met afbeeldingen te werken en die weer te geven, zie onderstaande code.
TODO: verander de url voor de url van een andere afbeelding'''
# Import een library die afbeeldingen kan inladen
from skimage.io import imread

# Import een library die afbeeldingen kan weergeven
import matplotlib.pyplot as plt

# Sla de url van de afbeelding op als een string
image_url = 'https://helpx.adobe.com/content/dam/help/en/photoshop/using/convert-color-image-black-white/jcr_content/main-pars/before_and_after/image-before/Landscape-Color.jpg'

# Laad de afbeelding (met skimage.io.imread)
afbeelding = imread(image_url)

# Laat de afbeelding zien (met matplotlib.pyplot.imshow en matplotlib.pyplot.show)
plt.imshow(afbeelding)
plt.show()

#BONUS: kan je de gemiddelde pixel waarde vinden van de afbeelding met numpy mean?
print('Gemiddelde is', np.mean(afbeelding))
```
