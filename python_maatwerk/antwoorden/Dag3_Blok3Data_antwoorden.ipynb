```python
''' Opdracht 5.
- Maak een DF (boodschappen_df) van de lijsten en laat het zien'''
data = [['spinazie', 'peer', 'water'], 
        ['aubergine', 'appel', 'limonade'],
        ['artisjok', 'meloen', 'ginger ale']]
columns = ['groeten', 'fruit', 'drank']
# Antwoord hier:
boodschappen_df = pd.DataFrame(data, columns=columns)
boodschappen_df
```
