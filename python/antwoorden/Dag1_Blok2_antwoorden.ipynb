```python
''' Deze code mag je negeren en hoef je niet te begrijpen; het is om je te helpen met checks en tips'''
def tip(index, items=['Gebruik het = teken om een waarde aan een variabele toe te wijzen.', 'Gebruik + om twee variabelen op te tellen.', 'Gebruik - om een getal van een ander af te trekken.', 'Gebruik * om twee getallen te vermenigvuldigen.', 'Gebruik / voor delen.', 'Gebruik // voor vloerdeling.', 'Gebruik % om de restwaarde van een deling te vinden.', 'Gebruik ** om een getal tot een bepaalde macht te verheffen.', 'Gebruik input() om gebruikersinput te krijgen en int() om het naar een integer om te zetten.', 'Gebruik print() om een waarde te tonen.', 'Gebruik # voor een enkele regel commentaar.', 'Gebruik + om strings samen te voegen.', 'Gebruik int() om een string naar een integer om te zetten.', 'Gebruik len() om de lengte van een string te bepalen.', 'Gebruik len() om de lengte van een string te bepalen.', 'Gebruik input() om gebruikersinput te vragen', 'Gebruik input() om gebruikersinput te vragen. Strings hoeven niet geconverteerd te worden.', 'Gebruik de code van opdracht 16.', 'Gebruik print() om een bericht naar de gebruiker te printen', 'Gebruik int(input("Vraag aan de gebruiker: ")) om een string naar een integer om te zetten.', 'Gebruik "Hallo " + naam en print']): return items[index-1] if type(index) == int and 1 <= index < len(items) + 1 else 'Er is geen tip voor deze opdracht'
def check(input, help): return 'Goed: je antwoord klopt volgens deze check' if input else help.replace('Fout:', 'Het antwoord klopt nog niet helemaal:')
```

```python
''' Opdracht 1: Variabelen toewijzen'''
# TODO: Maak een variabele genaamd `a` en wijs de waarde 5 toe.
a = 5  # Jouw antwoord

# Belangrijk: Op deze manier laat je een tip zien (1 staat voor opdracht 1)
print(tip(1))

# Deze regel hoef je niet te begrijpen, maar controleert of je antwoord goed is
check(a == 5, "Fout: de waarde van a is niet correct")
```
