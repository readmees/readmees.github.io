```python
''' Opdracht 13: Dictionary keys en values '''
# TODO: Maak een lijst van alle sleutels in de dictionary `persoon` en een lijst van alle waarden. Sla de lijsten op als `sleutels` en `waarden`.
persoon = {"naam": "Alida", "leeftijd": 31}
sleutels = list(persoon.keys())  # Jouw antwoord
waarden = list(persoon.values())  # Jouw antwoord
# Tip: Gebruik de .keys() en .values() methodes van een dictionary.
check(sleutels == ["naam", "leeftijd"] and waarden == ["Alida", 31], "Fout: de lijsten zijn niet correct")
```

```python
''' Opdracht 14: List slicing '''
# TODO: Maak een lijst genaamd `nummers` met de getallen 1 t/m 10. Gebruik slicing om de eerste vijf elementen te selecteren en sla deze op in `eerste_helft`.
nummers = list(range(1, 11)) # Dit betekend: een lijst van het bereik van 1 tot 11
eerste_helft = nummers[:5]  # Jouw antwoord
# Tip: Gebruik slicing met [startindex:eindindex] om delen van een lijst te selecteren.
check(eerste_helft == [1, 2, 3, 4, 5], "Fout: de eerste helft is niet correct geselecteerd")
```

```python
''' Opdracht 15: List slicing en stappen '''
# TODO: Maak een lijst `nummers` van 1 tot en met 10. Gebruik slicing om elke tweede waarde uit de lijst te halen en sla dit op in `elke_tweede`.
nummers = list(range(1, 11)) # Dit betekend: een lijst van het bereik van 1 tot 11
elke_tweede = nummers[::2]  # Jouw antwoord
# Tip: Gebruik slicing met [start:stop:stap] om elke tweede waarde te selecteren.
check(elke_tweede == [1, 3, 5, 7, 9], "Fout: de selectie van elke tweede waarde is niet correct")
```
